<h3>2. Sort based on second parameter in 2d list:</h3>
<p>This function is used in hashFunction (in City Class)</p>

```python
def sortSecond(val): 
    return val[1]
```

<h3>3. Node Class </h3>
<p>A class to store just the shape of the city: Each Node can be space or block</p>

```python
class Node :
    def __init__(self, nodeType) :
        self.type = nodeType #types -> b: block, s: space
```

<h3>4. Obj Class </h3>
<p>This class is implemented to store ambulance(type = a), patients(type = p), and hospitals(type = h)</p>
<p>Per every entity we store its <b>coordinates, type, and capacity (for hospital can be greater than 0)</b>. We have also another attribute called <b>realType</b> which is important just for hospitals, because when ambulance or patients come into hospital and then get out, we must be able to restore the previous value of that cell which was space or hospital.</p>

```python
class Obj:
    def __init__(self, x, y, nodeType, capacity = 0) : 
        self.x = x
        self.y = y
        self.type = nodeType
        self.realType = nodeType
        self.capacity = capacity
    def __lt__(self, other):
        return False
```

<h1 style = "text-align: center; margin-top: 50px">City: Main Class</h1> 
<p>City class consists of everything we need to solve the the problem, including bfs, ids, heuristics  (main functions) and some primitive functions to implement main functions.</p>
<br>
<h2>Modeling:</h2>
<h3>States : </h3>
<p>Every State is a list with these values : </p>
<p style="text-indent :2em;">[0] : a list of Obj which hold the positions of <b>a, h, and p</b> in current situation of city</p>
<p style="text-indent :2em;">[1] : a hash string for this situation which is generated by <b>getHash</b> function for states comparison and is unique for every state.</p>
<p style="text-indent :2em;">[2] : The path the ambulance has traversed</p>
<p style = "font-size: 16px;">So the <b>Initial state</b> is implemented from the input *.txt file with primary positions of ambulance, patients, and hospitals.</p>
<p style = "font-size: 16px;"><b>Goal state</b> is When we have no patient in city which means we have no patient in state[0] list (or even in state[1]: hash of current state)</p>
<p style = "font-size: 16px;"><b>Action</b> means a change in state (a, p, or h), which means a different state with different hash string. for example if we move ambulance up, down, right, or left; if this movement is feasible, then we would have a new state.</p>
<br>

<h2>Primary Functions:</h2>
<br>

<h3>__init__</h3>
<p style="text-indent :2em;">Sets initial state, and a copy of it (we use this copy to restore current state when we change current state)</p>

<h3>getTableNodes</h3>
<p style="text-indent :2em;">Makes list of city nodes: blocks and non-blockes (used only in __init__)</p>

<h3>getHash</h3>
<p style="text-indent :2em;">Generates a unique hash string for every unique state, but how?</p>
<p style="text-indent :2em;">First we store the coordinates of every a, p, and h, then we sort them based on x and y. Finally we put x, y, and type of each node of types: p, h, and a respectively. So for every different positions of these entities (a,p, and h) we will have different hash.</p>
<p style="text-indent :2em;">For example: p(s): [1,1], [4,5]</p>
<p style="text-indent :2em;">h(s): [1,2], [6,0]</p>
<p style="text-indent :2em;">a: [3,2]</p>
<p style="text-indent :2em;">hash string would be: "1|1|p|4|5|p|#1|2|h|6|0|h|#3|2|a"</p>
<p style="text-indent :2em;"><b>note:</b> we add a path length to this hash string in ids to generate unique hash for unique path lengthes</p>

<h3>getInitialState</h3>
<p style="text-indent :2em;">Just to make initial state...</p>

<h3>getNumOfPatients</h3>
<p style="text-indent :2em;">Just to return remaining patients (used to check goal state).</p>

<h3>currentAmbulanceCoord</h3>
<p style="text-indent :2em;">Returns position of a in city.</p>

<h3>getNewDir</h3>
<p style="text-indent :2em;">The directions we used in algorithms and the real directions are different. This function converts alg directions to real directions.</p>

<h3>getNewXY</h3>
<p style="text-indent :2em;">Takes x, y , and direction and return the new x and new y which is made by the direction.</p>

<h3>getCurrentObj</h3>
<p style="text-indent :2em;">Takes x and y and return 's' or 'b' if this cell is empty or block, or returns Obj of this cell if its type is: a, p, or h</p>

<h3>checkLock</h3>
<p style="text-indent :2em;">An Optimization function to check whether it is possible to reach the goal from this state or we would lock because a patient is not moveable anymore.</p>

<h3>checkDirectionAndState</h3>
<p style="text-indent :2em;">This function checks whether the new state is feasible or not, if the new position of ambulace is a block, or if checkLock function returns true, or if moving patient is not possible then returns -1, else makes new state and returns it.</p>

<h3>isRepetitiveState</h3>
<p style="text-indent :2em;">checks repetitive states in algorithms with a <b>set of states</b> stored in <b>self.qSet</b> .</p>

<h3>getCopy</h3>
<p style="text-indent :2em;">Copy.deepcopy is a slow function, so we write a new funcion to do it faster.</p>
 

```python
import copy
class City :
    def __init__(self, table) :
        self.table = self.getTableNodes(table)
        self.bfsList = []
        self.currentState = self.getInitialState(table) #[0] state of a, p , and h.     [1] hash.       [2] path
        self.currentStateCopy = copy.deepcopy(self.currentState)
        self.qSet = set()
    def getTableNodes(self, table) :
        nodes = []
        for i in range(0, len(table)):
            currentNodes = []
            for j in range(0, len (table[i])): 
                if (table[i][j] == '#'):
                    currentNodes.append(Node('b'))
                else :
                    currentNodes.append(Node('s'))
            nodes.append(currentNodes)
        return nodes
      
    def getHash(self, objList): 
        currentHash = '' 
        currentP = []
        currentH = []
        currentA = []
        currentType = 'p'
        
        for i in objList: 
            if (i.type == 'p') :
                currentP.append([i.x, i.y])
            if (i.type == 'h') :
                currentH.append([i.x, i.y])
            if (i.type == 'a') :
                currentA.append([i.x, i.y]) 
        currentP.sort(key = sortSecond)
        currentH.sort(key = sortSecond)
        currentA.sort(key = sortSecond) 
        currentP.sort()
        currentH.sort()
        currentA.sort()  
        for i in currentP:
            currentHash += str(i[0]) + "|" + str(i[1]) + "|" + 'p' + "|" 
        currentHash += "#"
        for i in currentH:
            currentHash += str(i[0]) + "|" + str(i[1]) + "|" + 'h' + "|" 
        currentHash += "#"
        for i in currentA:
            currentHash += str(i[0]) + "|" + str(i[1]) + "|" + 'a' + "|"
        return currentHash
    def getInitialState(self, table) : 
        currentState = []
        currentState0 = []
        x = 0
        y = 0
        for i in range(0, len(table)):
            for j in range(0, len (table[i])): 
                if (table[i][j] == 'P'):
                    currentState0.append(Obj(i, j, 'p'))
                elif (table[i][j] == '0' or table[i][j] == '1' or table[i][j] == '2' or table[i][j] == '3' or table[i][j] == '4' or table[i][j] == '5') : 
                    capacity = int(table[i][j])
                    currentState0.append(Obj(i, j, 'h', capacity))
                elif (table[i][j] == 'A') : 
                    currentState0.append(Obj(i, j, 'a'))
                    x = i
                    y = j
        currentHash = self.getHash(currentState0)
        currentPath = '' 
        currentState.append(currentState0)
        currentState.append(currentHash)
        currentState.append(currentPath)
        currentState.append(x)
        currentState.append(y)        
        return currentState
    def getNumOfPatients(self) :
        ans = 0
        for i in range(0, len(self.currentState[0])) :
            if self.currentState[0][i].type == 'p':
                ans += 1
        return ans
    
    def currentAmbulanceCoord(self) : 
        for i in self.currentState[0] :
            if i.type == 'a':
                return [i.x, i.y]
        return [-1, -1]
    def getNewDir(self, direction) : 
        if (direction == 'u') :
            return 'r'
        if (direction == 'r') :
            return 'd'
        if (direction == 'd') :
            return 'l'
        if (direction == 'l') :
            return 'u'
    def getNewXY(self, x, y, direction) : 
        if (direction == 'u') : 
            return [x, y + 1]
        if (direction == 'r') : 
            return [x + 1, y]
        if (direction == 'd') : 
            return [x, y - 1]
        if (direction == 'l') : 
            return [x - 1, y]
    
    def getCurrentObj(self, x, y) :
        if (self.table[x][y].type == 'b') : 
            return 'b'
        for i in self.currentState[0] : 
            if (i.x == x and i.y == y) : 
                return i
        return 's'
    def setNewAmbulance(self, x, y, newX, newY) :
        for i in range(0, len(self.currentState[0])) : 
            if (self.currentState[0][i].x == x and self.currentState[0][i].y == y and self.currentState[0][i].type == 'a' and self.currentState[0][i].realType != 'h') :
                self.currentState[0].pop(i)
                break
            elif (self.currentState[0][i].x == x and self.currentState[0][i].y == y and self.currentState[0][i].type == 'a' and self.currentState[0][i].realType == 'h') :
                self.currentState[0][i].type = 'h'
                break
        tempObj = self.getCurrentObj(newX, newY)
        if (tempObj == 's') :
            newObj = Obj(newX, newY, 'a')
            self.currentState[0].append(newObj)
        elif (tempObj != 'b' and ((tempObj.type == 'h') or (tempObj.type == 'p'))) : 
            for i in range(0, len(self.currentState[0])) : 
                if (self.currentState[0][i].x == newX and self.currentState[0][i].y == newY) :
                    self.currentState[0][i].type = 'a'
                    break
                
    def setNewPatient(self, newX, newY) : 
        for i in range(0, len(self.currentState[0])) : 
            if (self.currentState[0][i].x == newX and self.currentState[0][i].y == newY and self.currentState[0][i].type == 'h' and self.currentState[0][i].capacity > 0) :
                self.currentState[0][i].capacity -= 1
                return
            elif (self.currentState[0][i].x == newX and self.currentState[0][i].y == newY and self.currentState[0][i].type == 'h' and self.currentState[0][i].capacity <= 0) :
                self.currentState[0][i].type = 'p'
                return
        newObj = Obj(newX, newY, 'p')
        self.currentState[0].append(newObj)
        return
     
    def checkLock(self,x, y, direction) : 
        ans = 0
        if (self.table[x][y+1].type == 'b') : 
            ans += 1
        if (self.table[x+1][y].type == 'b') : 
            ans += 1
        if (self.table[x][y-1].type == 'b') : 
            ans += 1
        if (self.table[x-1][y].type == 'b') : 
            ans += 1
        if (ans < 2):
            return False
        elif (ans >= 2) : 
            if (self.table[x][y+1].type == 'b' and direction == 'u') : 
                return True
            if (self.table[x+1][y].type == 'b' and direction == 'r') : 
                return True
            if (self.table[x][y-1].type == 'b' and direction == 'd') : 
                return True
            if (self.table[x-1][y].type == 'b' and direction == 'l') : 
                return True
            else:
                return False            
    def checkDirectionAndState(self, x, y, direction):
        [newX,newY] = self.getNewXY(x, y, direction)
        if (self.table[newX][newY].type == 'b') : 
            return -1
        currentObj = self.getCurrentObj(newX, newY)
        if (currentObj == 's' or currentObj.type == 'h') :
            self.setNewAmbulance(x, y, newX, newY)
        elif (currentObj.type == 'p') :
            [newNewX, newNewY] = self.getNewXY(newX, newY, direction)
            if (self.table[newNewX][newNewY].type == 'b') : 
                return -1
            currentObj = self.getCurrentObj(newNewX, newNewY)
            if (currentObj != 's' and currentObj.type == 'p') :
                return -1
            elif (currentObj == 's' and self.checkLock(newNewX, newNewY, direction)) : 
                return -1
            elif (currentObj != 's' and currentObj.type == 'h' and currentObj.capacity <= 0 and self.checkLock(newNewX, newNewY, direction)) : 
                return -1
            self.setNewAmbulance(x, y, newX, newY)
            self.setNewPatient(newNewX, newNewY)
        self.currentState[1] = self.getHash(self.currentState[0])
        self.currentState[2] = self.currentState[2] + self.getNewDir(direction)
        self.currentState[3] = newX
        self.currentState[4] = newY
        return 1
        
    def isRepetitiveState(self, length = '') : 
        return ((self.currentState[1] + str(length)) in self.qSet)
    
    def getCopy(self, state) : 
        copy = []
        state0 = []
        for i in state[0] :
            obj = Obj(i.x, i.y, i.type, i.capacity)
            state0.append(obj)
        state1 = state[1]
        state2 = state[2]
        state3 = state[3]
        state4 = state[4]
        copy.append(state0)
        copy.append(state1)
        copy.append(state2)
        copy.append(state3)
        copy.append(state4)
        return copy
```

<h1 style = "text-align : center">BFS (Breathe First Search)</h1>
<div style = "align-items: center">
    <img src = "img3.png">
</div>
<h2>Algorithm:</h2>
<p>First of all we will start from the <b>root node : (initial state)</b>, then per every node in graph (in our problem: state), <b>we will visit all of its children</b>, and per every feasible child we will append its state in a q named <b>q</b>, then for next iterations, we pop from head of this queue and visit its children again and so on.</p>
<p>We will finish iteration in these ciscumstances: </p>
<p style="text-indent :2em;">1. Number of patients equals to 0 which means we have found the correct path</p>
<p style="text-indent :2em;">2. Length of queue equals to 0 which means there are no valid path.</p>
<br>
<p>For every state we have 4 possibility: up, down, right, and left. and to check feasibility of these possibilities we use two functions: </p>
<p style="text-indent :2em;">1. checkDirectionAndState() which checkes whether new state is possible or not. </p>
<p style="text-indent :2em;">2. isRepetitiveState() which checkes whether this state is repetitive or not. <b>Checking is done with hash strings</b> (we store every visited state in a set to reduce time consumption of checking).</p>
<br>
<p>if condition 1 is correct: we should restore previous state from copyState and increament <b>states</b> which holds total states</p>
<p>If everything is good, it means this state is feasible and we add this state into q and increament <b>uStates</b> which holds unique states</p>
<p>Ultimately if we find a path, we will return: <b>path, states, and uStates</b></p>
<br>
<h2>Advantages : </h2>
<br>
<p>Completeness : because we check every possible state.</p>
<p>Optimality : because we first check pathes with shorter length (1,2,3,...) so it will result optimality.</p>
<h2>Disadvantages: </h2>
<p>Number of states : we check every possible state in a certain length of path (though I optimized it with <b>checkLock()</b> ).</p>
<p>Complexity of space : because we must save every state we have seen which means all possible states.</p>



```python
def bfsSolution(self) :
        q = []
        start = True
        uStates = 1 #unique states
        states = 1 #total states
        reachEnd = False
        while (reachEnd == False) :
            if (start) : 
                start = False
            else :
                self.currentState = q[0]
                self.currentStateCopy = self.getCopy(self.currentState)
                q.pop(0)
                # self.qSet.remove(self.currentState[1])
            path = self.currentState[2]
            # print("path: ", path)
            # print("patients: ", self.getNumOfPatients())
            # print()
            [x, y] = [self.currentState[3], self.currentState[4]]
            
            if (self.getNumOfPatients() == 0 or (len(q) == 0 and uStates > 1)) : 
                reachEnd = True
                return [path, uStates, states]
            
            
            uResult = self.checkDirectionAndState(x, y, 'u')
            if (uResult == 1 and self.isRepetitiveState() == False) : 
                q.append(self.currentState)
                self.qSet.add(self.currentState[1])
                uStates += 1
            if (uResult != -1) :
                states += 1
                self.currentState = self.getCopy(self.currentStateCopy)
            
            rResult = self.checkDirectionAndState(x, y, 'r')
            if (rResult == 1 and self.isRepetitiveState() == False) : 
                q.append(self.currentState)
                self.qSet.add(self.currentState[1])
                uStates += 1
            if (rResult != -1) :
                states += 1
                self.currentState = self.getCopy(self.currentStateCopy)
            
            dResult = self.checkDirectionAndState(x, y, 'd')
            if (dResult == 1 and self.isRepetitiveState() == False) : 
                q.append(self.currentState)
                self.qSet.add(self.currentState[1])
                uStates += 1
            if (dResult != -1) :
                states += 1
                self.currentState = self.getCopy(self.currentStateCopy)
                
            lResult = self.checkDirectionAndState(x, y, 'l')
            if (lResult == 1 and self.isRepetitiveState() == False) : 
                q.append(self.currentState)
                self.qSet.add(self.currentState[1])
                uStates += 1
            if (lResult != -1) :
                states += 1
                path = self.currentState[2]
                # self.currentState = copy.deepcopy(self.currentStateCopy)
                
City.bfsSolution = bfsSolution
```

<h1 style = "text-align : center">IDS (Iterative deepening search)</h1>
<div style = "align-items: center">
    <img src = "img5.png" style = "width: 80%; margin-top: 50px;">
</div>
<h2>Algorithm:</h2>
<p style="text-indent :2em;">1. Check the root</p>
<p style="text-indent :2em;">2. Do a DFS searching for a path of length 1</p>
<p style="text-indent :2em;">3. If there is no path of length 1, do a DFS searching for a path of length 2</p>
<p style="text-indent :2em;">4. If there is no path of length 2, do a DFS searching for a path of length 3...</p>
<p style="text-indent :2em;">5. If you find the answer stop iteration and return result</p>

<p><b>idsSolution()</b> is the function which adds the length of DFS gradually and call <b>dls()</b> to find the solution. It also count the number of unique states in every iteration ans store it in <b>uStates</b> variable.</p>
<p style="text-indent :2em;"><b>note: </b>after each iteration of a certain length, we clear the <b>qSet</b> to be able to start from new ids length.</p>

<br>
<br>
<p><b>dls()</b> is the function that performs dfs with a certain dfs length : </p>
<div style = "align-items: center">
    <img src = "img4.png">
</div>
<p style="text-indent :2em;">For every state if first child's state is feasible, then call dls() for this child. So it will continue a path until it meet a constraint or correct path.</p>
<p style="text-indent :2em;">Like BFS we use these two functions to check feasibility: .</p>
<p style="text-indent :4em;">1. checkDirectionAndState() which checkes whether new state is possible or not. </p>
<p style="text-indent :4em;">2. isRepetitiveState() which checkes whether this state is repetitive or not. <b>Checking is done with hash strings *</b> (we store every visited state in a set to reduce time consumption of checking).</p>
<p style="text-indent :4em;"><b>* Hash string contains the path length at the end of funcion getHash(), because we should continue a repetitive state if current length is less that previous one.</b></p> 
<p>Finally we will return path if we found the correct one or false otherwise. The number of total states is also calculates in <b>states</b> which increaments in every possible iteration.</p>
<br>

<h2>Advantages : </h2>
<br>
<p>Completeness : because we check every possible state in every length.</p>
<p>Optimality : because we first check pathes with shorter length (1,2,3,...) so it will result optimality.</p>
<h2>Disadvantages: </h2>
<p>Number of states and time complexity : we check every possible state <b>multiple times because of the increament of length</b> though I optimized it with <b>checkLock()</b> .</p>


```python
def dls(self, state, depth, path, states) :
    self.currentState = self.getCopy(state)
    pathLength = len(path)
    if (self.getNumOfPatients() == 0) :
        return  [path, states]

    if (depth <= 0) : 
        return  [False, states]
    [x, y] = [self.currentState[3], self.currentState[4]]

    uResult = self.checkDirectionAndState(x, y, 'u')
    if (uResult != -1) :
        states += 1
    if (uResult == 1 and self.isRepetitiveState(pathLength) == False) : 
        res = self.dls(self.currentState, depth - 1, path + self.getNewDir('u'), states)
        states = res[1]
        if (res[0] != False) : 
            return res
        self.qSet.add(self.currentState[1] + str(pathLength))
    if (uResult != -1) :
        self.currentState = self.getCopy(state)

    rResult = self.checkDirectionAndState(x, y, 'r')
    if (rResult != -1) :
        states += 1
    if (rResult == 1 and self.isRepetitiveState(pathLength) == False) :
        res = self.dls(self.currentState, depth - 1, path + self.getNewDir('r'), states)
        states = res[1]
        if (res[0] != False) : 
            return res
        self.qSet.add(self.currentState[1] + str(pathLength))
    if (rResult != -1) :
        self.currentState = self.getCopy(state)

    dResult = self.checkDirectionAndState(x, y, 'd')
    if (dResult != -1) :
        states += 1
    if (dResult == 1 and self.isRepetitiveState(pathLength) == False) : 
        res = self.dls(self.currentState, depth - 1, path + self.getNewDir('d'), states)
        states = res[1]
        if (res[0] != False) : 
            return res
        self.qSet.add(self.currentState[1] + str(pathLength))
    if (dResult != -1) :
        self.currentState = self.getCopy(state)

    lResult = self.checkDirectionAndState(x, y, 'l')
    if (lResult != -1) :
        states += 1
    if (lResult == 1 and self.isRepetitiveState(pathLength) == False) : 
        res = self.dls(self.currentState, depth - 1, path + self.getNewDir('l'), states)
        states = res[1]
        if (res[0] != False) : 
            return res
        self.qSet.add(self.currentState[1] + str(pathLength))
    if (lResult != -1) :
        self.currentState = self.getCopy(state)

    return [False, states]
def idsSolution (self) : 
    uStates = 0
    states = 0
    for i in range(0, 1000):    
        temp = len(self.qSet)
        uStates = temp
        self.qSet.clear()
        currentDls = self.dls(self.currentState, i, '', 0)
        states += currentDls[1]
        if currentDls[0] != False :
            temp = len(self.qSet)
            uStates += temp
            return [currentDls[0], uStates, states]
    return "no path"
City.dls = dls
City.idsSolution = idsSolution
```

# City Class 3 (heuristic Functions)

```python
def heuristic1Func(self) : 
    hN = 0
    currentHash = '' 
    currentP = []
    currentH = []
    currentA = []
    currentType = 'p'
    disAP = 0
    disPH = 0
    for i in self.currentState[0]: 
        if (i.type == 'p') :
            currentP.append([i.x, i.y])
        if (i.realType == 'h') :
            currentH.append([i.x, i.y])
        if (i.type == 'a') :
            currentA = [i.x, i.y]
    for i in currentP : 
        newDisAP = abs(currentA[0] - i[0]) + abs(currentA[1] - i[1])
        if (newDisAP < disAP or disAP == 0) : 
            disAP = newDisAP
        disPH = 0
        for j in currentH : 
            newDisPH = abs(j[0] - i[0]) + abs(j[1] - i[1])
            if ((newDisPH < disPH or disPH == 0) and (newDisPH > 0)) : 
                disPH = newDisPH
        hN += disPH
    hN += disAP
    hN += len(self.currentState[2])
    return hN

def heuristic2Func(self) : 
    hN = 0
    currentHash = '' 
    currentP = []
    currentA = []
    currentType = 'p'
    disAP = 0
    disPH = 0
    for i in self.currentState[0]: 
        if (i.type == 'p') :
            currentP.append([i.x, i.y])
        if (i.type == 'a') :
            currentA = [i.x, i.y]
    for i in currentP : 
        newDisAP = abs(currentA[0] - i[0]) + abs(currentA[1] - i[1])
        if (newDisAP < disAP or disAP == 0) : 
            disAP = newDisAP
    hN += disAP + len(currentP)
    hN += len(self.currentState[2])
    return hN 

def getCurrentHN(self, hid) :
    if (hid == 1) : 
        return self.heuristic1Func()
    elif (hid == 2) : 
        return self.heuristic2Func() 
        # return 0

def heuristicSolutuion(self, hid) :
    q = []
    start = True
    uStates = 1 #unique states
    states = 1 #total states
    reachEnd = False
    while (reachEnd == False) :
        if (start) : 
            start = False
        else :
            index = 0
            counter = 0
            hn = 100000000
            for i in q :
                if (i[0] < hn) :
                    index = counter
                    hn = i[0]
                counter += 1
            self.currentState = q[index][1]
            self.currentStateCopy = self.getCopy(self.currentState)
            q.pop(index)
        path = self.currentState[2]
        [x, y] = [self.currentState[3], self.currentState[4]]

        if (self.getNumOfPatients() == 0 or (len(q) == 0 and uStates > 1)) : 
            reachEnd = True
            return [path, uStates, states]

        uResult = self.checkDirectionAndState(x, y, 'u')
        if (uResult == 1 and self.isRepetitiveState() == False) : 
            hN = self.getCurrentHN(hid)
            q.append([hN, self.currentState])
            self.qSet.add(self.currentState[1])
            uStates += 1

        if (uResult != -1) :
            self.currentState = self.getCopy(self.currentStateCopy)
            states += 1


        rResult = self.checkDirectionAndState(x, y, 'r')
        if (rResult == 1 and self.isRepetitiveState() == False) : 
            hN = self.getCurrentHN(hid)
            q.append([hN, self.currentState])
            self.qSet.add(self.currentState[1])
            uStates += 1

        if (rResult != -1) :
            self.currentState = self.getCopy(self.currentStateCopy)
            states += 1


        dResult = self.checkDirectionAndState(x, y, 'd')
        if (dResult == 1 and self.isRepetitiveState() == False) : 
            hN = self.getCurrentHN(hid)
            q.append([hN, self.currentState])
            self.qSet.add(self.currentState[1])
            uStates += 1

        if (dResult != -1) :
            self.currentState = self.getCopy(self.currentStateCopy)
            states += 1


        lResult = self.checkDirectionAndState(x, y, 'l')
        if (lResult == 1 and self.isRepetitiveState() == False) : 
            hN = self.getCurrentHN(hid)
            q.append([hN, self.currentState])
            self.qSet.add(self.currentState[1])
            uStates += 1

        if (lResult != -1) :
            path = self.currentState[2]
            states += 1

City.heuristic1Func = heuristic1Func
City.heuristic2Func = heuristic2Func
City.getCurrentHN = getCurrentHN
City.heuristicSolutuion = heuristicSolutuion
```

<h1 style = "text-align: center">Calling search functions : </h1>
<p>Every function calls its relative function and calculate time consumption with <b>time.time()</b>, then return <b>path, total states, unique states, and duration of process in seconds</b></p>

```python
import time
def bfs(i) : 
    if (i == 1) : 
        city = City(l1)
    elif (i == 2) : 
        city = City(l2)
    elif (i == 3) : 
        city = City(l3)
        
    start = time.time()
    response = city.bfsSolution()
    end = time.time()
    
    duration = end - start
    path = response[0]
    uStates = response[1]
    states = response[2]
    
    return [path, states, uStates, duration]
    
    
def ids(i) :    
    if (i == 1) : 
        city = City(l1)
    elif (i == 2) : 
        city = City(l2)
    elif (i == 3) : 
        city = City(l3)
        
    start = time.time()
    response = city.idsSolution()
    end = time.time()
    
    duration = end - start
    path = response[0]
    uStates = response[1]
    states = response[2]
    
    return [path, states, uStates, duration]
    
def heuristic(hid, i) :    
    if (i == 1) : 
        city = City(l1)
    elif (i == 2) : 
        city = City(l2)
    elif (i == 3) : 
        city = City(l3)
        
    start = time.time()
    response = city.heuristicSolutuion(hid)
    end = time.time()
    
    duration = end - start
    path = response[0]
    uStates = response[1]
    states = response[2]
    
    return [path, states, uStates, duration]
```

<h1 style = "text-align: center">Execution : </h1>
<p><b>getTableRows()</b> makes all the required data and average execution time from 3 input files, and return them as <b>res1, res2, and res3</b></p>

```python
import pandas as pd

def getAverage(lst): 
    return sum(lst) / len(lst) 

def getTableRows(funcType, hType = 1) : 
    res1 = []
    res2 = []
    res3 = []
    times1 = []
    times2 = []
    times3 = []
    bfs3Times = []
    for i in range(0,1) : 
        if (funcType == "bfs") : 
            res1 = bfs(1)
            res2 = bfs(2)
            res3 = bfs(3)
        elif (funcType == "ids") : 
            res1 = ids(1)
            res2 = ids(2)
            res3 = ids(3)
        elif (funcType == "heuristic") : 
            res1 = heuristic(hType, 1)
            res2 = heuristic(hType, 2)
            res3 = heuristic(hType, 3)
        times1.append(res1[3])
        times2.append(res2[3])
        times3.append(res3[3])
    avgTime1 = getAverage(times1)
    avgTime2 = getAverage(times2)
    avgTime3 = getAverage(times3)
    res1[3] = avgTime1
    res2[3] = avgTime2
    res3[3] = avgTime3
    res1[0] = len(res1[0])
    res2[0] = len(res2[0])
    res3[0] = len(res3[0])
    return [res1, res2, res3]
```

```python
bfsResult = getTableRows("bfs")
idsResult = getTableRows("ids")
h1Result = getTableRows("heuristic", 1)
h2Result = getTableRows("heuristic", 2)
```
